# 데이터 불러오기

```python
import pandas as pd

df = pd.read_csv("C:/Users/atoz/Downloads/연관키워드.csv", encoding='cp949')
df
```

# 데이터 전처리

```python
# < 10인 걸 -> 9, ,를 ''으로 변환하는 함수
def clean_search_count(x):
    if isinstance(x, str):
        if '< 10' in x:
            return 9
        else:
            return int(x.replace(',', ''))  # 콤마 제거 후 정수로 변환
    return x

# PC와 모바일 검색수 전처리
df['월간검색수(PC)'] = df['월간검색수(PC)'].apply(clean_search_count)
df['월간검색수(모바일)'] = df['월간검색수(모바일)'].apply(clean_search_count)

# 전처리된 데이터로 총검색수 계산
df['총검색수'] = df['월간검색수(PC)'] + df['월간검색수(모바일)']
df
```
